```python
import pandas as pd
import  numpy as np
import datetime as dt
import matplotlib.pyplot as plt
```

# Visualizando patrones de series de tiempo

**DATOS**

```python
#carga
df  = pd.read_csv('Ti_data.csv', index_col = 0, parse_dates = True)
```

```python
#Plot
fig, ax = plt.subplots(1, 1, figsize  = (20,5))
df.plot(ax = ax, color = 'C1', legend = None)
ax.set_ylim(17.15, 22)
ax.set_ylabel('$C$')
ax.set_title('Internal Temperature $(T_i)$')
```

# Visualizando patrones diurnos

```python
#Seporando da y hora
#creamos nuevas columnas donde separaremos las fechas del tiempo
df['date'] = df.index.normalize()
df['time'] = df.index.time.astype(str)
df
```

```python
#Transformando el dataframe
#para esto se crea un nuevo dataset y el time o las horas seran columnas y el dinice describe la fecha

df_pivot = df.pivot(index = "date", columns = 'time', values = 'Ti')
df_pivot
```

```python
#Grafico del dataset pivot
fig, ax = plt.subplots(1, 1, figsize = (10,5))
df_pivot.T.plot(ax = ax, color = 'C2', legend = False)
ax.set_xlim([0,47])
ax.set_ylim(17.75, 24)
ax.set_ylabel('$C$')
ax.set_xlabel('$Tiempo$')
ax.set_title('Internal Temperature $(T_i)$')
```

# Transparencia Alpha 

```python
fig, ax = plt.subplots(1, 1, figsize = (10,5))
df_pivot.T.plot(ax = ax, color = 'C2', legend = False, alpha =0.2)
ax.set_xlim([0,47])
ax.set_ylim(17.75, 24)
ax.set_ylabel('$C$')
ax.set_xlabel('$Tiempo$')
ax.set_title('Internal Temperature $(T_i)$')
```

# Cambiando el horario de comienzo 

```python
#El dataset empieza a las 18:00 entonces pasaremos de las 00:00 a las 18
#18 horas son 36 medias horas (las medias horas es el periodo en el que se toma la temperatura)
df_roll = pd.DataFrame(index = df_pivot.index,
                      columns = np.roll(df_pivot.columns, -36),
                      data = np.roll(df_pivot.values, -36))

fig, ax = plt.subplots(1, 1, figsize =(15,6))
df_roll.T.plot(ax = ax, color = 'C3', alpha = 0.2, legend = False)
ax.set_xlim([0,47])
ax.set_ylim(17, 22)
ax.set_ylabel('$C$')
ax.set_title('Internal Temperature $(T_i)$');
```

# Visualizando patrones dirunos y semalaes

```python
day_colors = {'Monday':'C0',
              'Tuesday':'C1',
              'Wednesday':'C1',
              'Thursday': 'C1',
              'Friday' : 'C1',
              'Saturday' : 'C2',
              'Sunday' : 'C3'}

#Plot
fig, ax = plt.subplots(1, 1, figsize = (15, 6))

for day_name, color in day_colors.items():
    df_roll[df_roll.index.day_name() == day_name].T.plot(
    ax = ax, color = color, alpha = 0.2, legend = None)
    
ax.set_xlim([0,47])
ax.set_ylim(17.5, 22)

#Lineas verticales para definir los horarios de calefaccin
ax.axvline(30, alpha = 1, color = 'black', linestyle = '--')
ax.axvline(40, alpha = 0.5, color = 'black', linestyle = '--')
ax.axvline(45, alpha = 0.3, color = 'black', linestyle = '--')

ax.set_ylabel('$C$')
ax.set_title('Internal Temperatures $(T_i)$')

#Leyenda
from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color = 'C0', lw=2),
                Line2D([0], [0], color = 'C1', lw=2),
                Line2D([0], [0], color = 'C2', lw=2),
                Line2D([0], [0], color = 'C3', lw=2)]
ax.legend(custom_lines, ['Mon', 'Tue-Frid', 'Sat', 'Sun']);
```
