```python
import pandas as pd
import numpy as np
#ARResults nos devuelve la cantidad de retornos que podemos hacer
from statsmodels.tsa.ar_model import AR, ARResults
%matplotlib inline
```

```python
df = pd.read_csv('uspopulation.csv', index_col ='DATE', parse_dates = True)
#Frecuencia en meses
df.index.freq = 'MS'
df.tail()
```

# Grafico de la serie

```python
title = 'U.S Población Estimada'
ylabel = 'Pop. Est. (Thousands)'
xlabel = 'Años'
ax = df['PopEst'].plot(figsize = (12,9), title = title)
ax.autoscale(axis = 'x', tight= True)
ax.set(xlabel= xlabel, ylabel = ylabel);
```

# Datos de entrenamiento y prueba

```python
len(df)*0.8
```

```python
train = df.iloc[:84]
test = df.iloc[84:]
```

# Modelo AR(1)

```python
#No salgan los warning
import warnings
warnings.filterwarnings('ignore')
```

```python
#seleccionamos los datos del modelo
model = AR(train['PopEst'])
#maxlag = modelo de orden 1, method = metodo maxima verosimilitud
AR1fit = model.fit(maxlag = 1, method = 'mle')
print(f'lag: {AR1fit.k_ar}')
print(f'Coefficientes: \n {AR1fit.params}')
```

```python
#Realizando predicciones
#inicia donde termina el conjunto de entrenamiento
start = len(train)
#termina donde termina el test
end = len(train)+len(test)-1
predictions1 = AR1fit.predict(start = start, end = end, dynamic = False).rename('AR(1) Predicciones')
```

```python
predictions1
```

```python
test['PopEst'].plot(legend = True)
predictions1.plot(legend = True)
```

# Modelo de profundidad 2 AR(2)

```python
model = AR(train['PopEst'])
AR2fit = model.fit(maxlag = 2, method = 'mle')
print(f'lag: {AR2fit.k_ar}')
print(f'Coefficientes: \n {AR2fit.params}')
```

```python
start = len(train)
end = len(df)-1
predictions2 = AR2fit.predict(start = start, end = end,
                             dynamic = False).rename('AR(2) predicciones')
test['PopEst'].plot(legend = True)
predictions2.plot(legend = True)
predictions1.plot(legend = True)
```

# Statsmodels selecciona la profundidad

```python
model = AR(train['PopEst'])
#ic = considera hasta un n numero de retornos y devuelve la cantidad de retornos correctos
ARfit = model.fit(ic = 'bic')
print(f'Lags: {ARfit.k_ar}')
print(f'Coeficientes: {ARfit.params}')
```

```python
start = len(train)
end = len(df)-1
rename = 'AR(8) Predicciones'
prediccion8 = ARfit.predict(start = start, 
                             end = end, 
                             dynamic = False).rename(rename)
```

```python
test['PopEst'].plot(legend = True)
predictions1.plot(legend = True)
predictions2.plot(legend = True)
prediccion8.plot(legend = True)
```

# Evaluación del modelo

```python
from sklearn.metrics import mean_squared_error as mse
labels = ['AR(1)', 'AR(2)', 'AR(8)']
preds = [predictions1, predictions2, prediccion8]

for i in range(len(labels)):
    #Calculamos 3 errores entre el modelo y el test
    error = mse(test['PopEst'], preds[i])
    
    print(f'{labels[i]} Error: {error:11.8}')
```

```python
modls = [AR1fit, AR2fit, ARfit]
#devolviendo el AIC, para esto en el modls ponemos un .aic
for i in range (len(modls)):
    print(f'{labels[i]} AIC {modls[i].aic:6.5}')
```

# Predicciones en el futuro

```python
#Reajustamos el modelo ahora con todo el dataset
model = AR(df['PopEst'])
#Ajustamos el modelo
ARfit = model.fit(maxlag = 8, method = 'mle')
#Hacemos un forecast para 12 periodos despues (en este caso los periodos seran mes a mes)
fcast = ARfit.predict(start = len(df), end = len(df)+12, dynamic = False).rename('Forecast')

df['PopEst'].plot(legend= True)
fcast.plot(legend = True, figsize = (12,6))
```
