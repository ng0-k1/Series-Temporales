```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
df = pd.read_csv('airline_passengers.csv', index_col = 'Month', parse_dates = True)
```

```python
df.dropna(inplace = True)
```

```python
#Aqui definimos la frecuencia, que en este caso es mensual
df.index.freq = 'MS'
df.index
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
#Meses
span = 12
#Alpha se calcula dependiendo del span +1
alpha = 2/(span+1)

df['EWMA12'] = df['Thousands of Passengers'].ewm(alpha = alpha, adjust = False).mean()
#ajustamos el metodo con un fit y el smoothing es el nivel de suavizado
#Optimize  Calcula los parametros optimos del modelo
#fittdevalues devuelve los valores estimados del modelo
df['SES12'] = SimpleExpSmoothing(df['Thousands of Passengers']).fit(smoothing_level= alpha, 
                                                                    optimized =False).fittedvalues.shift(-1)
```

```python
df.head()
```

# Exponencial Doble

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
#Trend = tendecia, para este caso es aditiva suponiendo que la tendencia es lineal, en caso de que
#fuera exponencial (curva) entonces se usa multiplicativo.
#NOTA, RECORDAR QUE EL FIT ENTRENA Y EL FITTEDVALUES LOS GUARDA
df['DESadd12'] = ExponentialSmoothing(df['Thousands of Passengers'], trend ='add').fit().fittedvalues.shift(-1)
df.head()
```

```python
df[['Thousands of Passengers', 'EWMA12','DESadd12']].iloc[:24].plot(figsize = (12,6)).autoscale(axis ='x', tight = True)
```

```python
df['DESmul12'] = ExponentialSmoothing(df['Thousands of Passengers'], trend ='mul').fit().fittedvalues.shift(-1)
df.head()
```

# Suavizado exponencial triple
Aade soporte para la tendencia y la estacionalidad

```python
#Trend = Tendencia
#Seasonal = Estacionalidad parametro de la cantidad de periodos
df['TESadd12'] = ExponentialSmoothing(df['Thousands of Passengers'], 
                                      trend = 'add', 
                                      seasonal ='add', 
                                      seasonal_periods = 12).fit().fittedvalues
df.head()
```

```python
df['TESmul12'] = ExponentialSmoothing(df['Thousands of Passengers'], 
                                      trend = 'mul', 
                                      seasonal ='mul', 
                                      seasonal_periods = 12).fit().fittedvalues
df.head()
```

```python
df[['Thousands of Passengers', 'TESadd12', 'TESmul12']].iloc[:12].plot(figsize = (12,6));
```

```python
model_mult = df['TESmul12'] = ExponentialSmoothing(df['Thousands of Passengers'], 
                                      trend = 'mul', 
                                      seasonal ='mul', 
                                      seasonal_periods = 12).fit()
```

```python
forecast_mul = model_mult.forecast(36)
```

```python
df['Thousands of Passengers'].plot(figsize = (12,6))
forecast_mul.plot(label = "Forecast Multiplicative")
plt.legend(loc = 'upper left')
```
